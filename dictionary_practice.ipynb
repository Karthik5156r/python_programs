```python
#creating empty dict and doing operation
info={}
info["name"]="karthik"
info["age"]=19
print("dictionary",info)
```

```python
#creating dict from two list that one as key another one is value
list1=[1,2,3,4,5]
list2=[1,4,9,16,25]
d=dict(zip(list1,list2))
print(d)
```

```python
#creating duplicate list and making dictionary from that
lis__t=[1,2,2,3,3,4,4,4,4]
dic_t={}
for i in lis__t:
    dic_t[i]=lis__t.count(i)
print("numbers occuring:",dic_t)

```

```python
#iterating dictionaries using for loop
dic_t={"name":"karthik","age":19,"sport":"cricket"}
for keys,values in dic_t.items():
    print(f"{keys}:{values}")
```

```python
#creating dictionaries from list that key as list elements and value is its squares
lis_t=[1,2,3,4,5]
sqaure={x:x**2 for x in lis_t}
print(sqaure)
```
