```python
#coverting list characters to upper case
names=['karthik','jeeva','santhosh','yogesh']
for name in names:
    print(name.upper())
```

```python
#finding the length of characters in list
names=['karthik','jeeva','santhosh','yogesh']
for name in names:
    print(f"{names}->{len(name)} characters")
```

```python
#adding new elements in list using index numbers
num_list=[10,20,30,40,50]
index=int(input("enter the values(index):"))
values=int(input("enter the key:"))
num_list.insert(index,values)
print("new list:",num_list)
```

```python
#creating list and separate into odd and even numbers
numbers=list(range(1,101))
odd_numbers=[n for n in numbers if n%2!=0]
even_numbers=[n for n in numbers if n%2==0]
print("odd numbers:",odd_numbers)
print("even numbers:",even_numbers)
```
